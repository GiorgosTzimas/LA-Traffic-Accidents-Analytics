## <span style="color:red;">EDA</span>

#### Quick inspection of initial dataset focused only on Los Angeles

```python
sns.scatterplot(data=city, x="Start_Lng", y="Start_Lat", alpha=0.2, hue="Severity", palette="rocket_r")
```

```python
numerical_col = ["Severity", "Temperature(F)", "Humidity(%)", 
        "Wind_Speed(mph)", "Precipitation(in)"]

for col in numerical_col:
    plt.figure(figsize=(10,6))
    sns.histplot(city[col], kde=True, bins=50, color='blue')
    plt.title(f"Histogram plot of {col}")
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

    plt.figure(figsize=(8,4))
    sns.boxplot(x=city[col], color='green')
    plt.title(f"Boxplot of {col} to detect outliers")
    plt.xlabel(col)
    plt.show()
```

```python
categorical_col = ["Weather_Condition", "Amenity", "Bump", "Crossing", "Give_Way", 
        "Junction", "No_Exit", "Railway", "Roundabout", "Station", "Stop", "Traffic_Calming", "Traffic_Signal", 
        "Turning_Loop", "Sunrise_Sunset", "Civil_Twilight", "Nautical_Twilight"]

for col in categorical_col:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=col, data=city, palette='husl', order= city[col].value_counts().index,hue=col,legend=False)
    plt.title(f"Count of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```

```python
cols = ["ID", "Severity", "Start_Time", "Start_Lat", "Start_Lng", "Street", "Temperature(F)", "Humidity(%)", 
        "Wind_Speed(mph)", "Precipitation(in)", "Weather_Condition", "Amenity", "Bump", "Crossing", "Give_Way", 
        "Junction", "No_Exit", "Railway", "Roundabout", "Station", "Stop", "Traffic_Calming", "Traffic_Signal", 
        "Turning_Loop", "Sunrise_Sunset", "Civil_Twilight", "Nautical_Twilight"]
```
